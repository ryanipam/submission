# importing library for helping plot data

```python
import seaborn as sns 
import pandas as pd 
import matplotlib.pyplot as plt 
import matplotlib.dates 
from datetime import datetime 

%matplotlib inline

plt.style.use('seaborn')
```

# I put the data locally and use pandas for load

```python
data = pd.read_csv("dataset.csv")
```

```python
data.head()
```

# Plotting user_id using seaborn.distplot

```python
sns.distplot(data['user_id'], bins=50)
```

```python
# based from chart above, probability user for shoping dominated with user_id arount 80000-90000 and around 110000
```

# ploting product_id for knowing best selling product

```python
sns.distplot(data['product_id'], bins=50)
```

```python
#based from plot below, best seller product purchased is product with user_id < 200
```

# Joining user_id and product_id

```python
sns.jointplot(x='user_id',y='product_id',data=data, kind="hex")
```

```python
#Based on result join below, user_id which have dominant shop inclined to buy product which product_id under 200 (more precisely under 100)
```
